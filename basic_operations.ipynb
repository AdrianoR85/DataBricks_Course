```python
# Load the 'movies' table from the 'workspace.default' database into a Spark DataFrame and display it
df = spark.table("workspace.default.movies")
display(df)
```

```python
# Print the schema of the DataFrame to show column names and data types
df.printSchema()
```

```python
# Count the number of rows in the DataFrame
df.count()
```

```python
# Show all columns names in the DataFrame
df.columns
```

```python
# Generate summary statistics for each column in the DataFrame
display(df.describe())
```

```python
# Generate summary statistics for each column in the DataFrame
display(df.summary())
```

## Column Filtering

```python
# Select specific columns from the DataFrame
df_trimmed = df.select('title', "imdb_rating", "industry")
df_trimmed.show(3, truncate=False)
```

## Row Filtering

```python
# Show all movies released between 2000 and 2010
df_filtered = df.filter((df.release_year >= 2000) & (df.release_year <= 2010))
display(df_filtered)
```

```python
from pyspark.sql.functions import col

df_filtered = df.filter((col("release_year") >= 2000) & (col("release_year") <= 2010))
display(df_filtered)
```

```python
df_marvel = df.filter(col("studio") == 'Marvel Studios')
display(df_marvel)
```

```python
df.select("industry").distinct().show()
```

```python
df.select("language").distinct().show()
```

## Column Creating

```python
df.withColumn("profit", col("revenue") - col("budget")).show()
```

```python
df.withColumnRenamed("revenue", "total_revenue").show()
```
